#### Ex.2.1 Compare the number of participants of the marathon by using bar plot directly from Series.

#### Ex.2.2 Draw a blue histogram of the ages of participants

```r
# code for Ex.2.2

ggplot(data, aes(x=age))+ geom_histogram(fill="blue") 
```

#### Ex.2.3 Draw a green density plot of the ages of participants

```r
# code for Ex.2.3

ggplot(data, aes(x=age))+ geom_density(fill='green', alpha=0.3) + theme_bw()
#alpha gives the shade of the color higher means stronger lower means lighter
#Alpha means transparency
```

#### Ex.2.4 Draw histograms of the ages of male and female participants seperately

```r
# code for Ex.2.4
ggplot(data, aes(x=age,y= gender))+ geom_histogram(aes()) 
```

#### Ex.2.5 Draw a scatter to show the relationship between average time of splits and final time.

```r
# code for Ex.2.5
ggplot(data, aes(x=split, y=final)) + geom_point() + scale_color_gradient(low='blue',high='red')
```
